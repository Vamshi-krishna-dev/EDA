## 1) Testing The Salary Claim

**Research Question : Is the statement that fresh graduates in roles like Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer earn up to 2.5-3 lakhs true for the Given Data?**

**Approach**
1) Filter the dataset to include only the relevent job roles: Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer.
2) Calculate the average, minimum, maximum salary for this role
3) Perform a statistical test to compare the sample mean salary with the claim (2.5 - 3)lakhs

```python
from scipy import stats
```

```python
# Filter for relevant job roles
relevent_jobs = ['programming analyst', 'software engineer', 'hardware engineer','associate engineer']

# Convert roles to lowercase to match potential cases
data['Designation'] = data['Designation'].str.lower()

# Filter rows where designation is in the relevant job list
filtered_data = data[data['Designation'].isin(relevent_jobs)]
```

```python
# Check for missing or NaN values in salary data
filtered_data = filtered_data.dropna(subset = ['Salary'])
```

```python
# Calculate average salary
avg_salary = filtered_data['Salary'].mean()
min_salary = filtered_data['Salary'].min()
max_salary = filtered_data['Salary'].max()
```

```python
# Perform one-sample t-test, assuming population mean of 2.75 lakhs

Population_mean = 275000
t_stat,p_value = stats.ttest_1samp(filtered_data['Salary'],Population_mean)
```

```python
print("-----Salary Claim Test-----")
print(f"Average Salary: {avg_salary}, Min Salary: {min_salary}, Max Salary: {max_salary}")
print(f"T-test Statistic: {t_stat}, p-value: {p_value}")
```

### 2) RelationShip Between Gender and Specializaton

```python
#create a crosstab of gender and specialization
crosstab = pd.crosstab(data['Gender'],data['Specialization'])
```

```python
#Perform chi-square Test
chi2_stat, p_val, dof, expected = stats.chi2_contingency(crosstab)
print('-----------Gender and Specialization------------')
print(f"chi_square statistics: {chi2_stat},p_value:{p_val}")
```
