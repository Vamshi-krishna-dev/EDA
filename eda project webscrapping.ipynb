* The problem at hand is to gather data on various bikes from a website and analyze their attributes to understand the factors influencing their popularity and performance.



## Project Overview:
* The project aims to analyze data related to bikes, including their features such as price, ratings, cubic capacity, kilometers per liter, horsepower, and weight. The goal is to derive insights from the data and visualize various relationships between these features.


## Objective:

* To analyze the relationship between bike attributes (brand, price, engine specifications, etc.) and their ratings.
* To identify key factors influencing the popularity and performance of bikes.
* To provide insights for consumers, manufacturers, and enthusiasts in the motorcycle industry.
* With this problem statement, we aim to extract meaningful insights from the collected data to better understand the dynamics of the bike market and provide valuable information for various stakeholders.



```python
import re

import numpy as np,pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

import requests

from bs4 import BeautifulSoup
```

```python
All_Brands=['honda','royalenfield','hero','bajaj','yamaha',
   'tvs','kawasaki','bmw','ktm','triumph',
   'ampere','aprilia','ather','benelli','benling',
   'bgauss','bounce','bsa','ducati','evolet',
   'gemopai','harleydavidson','heroelectric','hop-electric','husqvarna',
   'indian','ivoomi','jawa','joyebike','keeway',
   'kinetic-green','lambretta','lectrix','matter','motoguzzi',
   'motomorini','norton','odysse','okaya','okinawa',
   'ola','pureev','qj-motor','revolt','simpleenergy',
   'suzuki','tork','ultraviolette','vespa','hero-vida',
   'yezdi','yo','zontes']
```

```python
Brand=[]

Price=[]

Rating_members=[]

Rating=[]

CC=[]

KMPL=[]

BHP=[]

Weight=[]

for i in All_Brands:
    url = f"https://www.bikewale.com/{i}-bikes/"
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    
    
    for i in soup.find_all("div",class_="o-brXWGL o-fznVqX o-fznVsN o-fznVmz o-fznVCp o-cpnuEd"):
        

            brand=i.find("h3", class_="o-jjpuv o-cVMLxW o-mHabQ o-fzpibK")
            if brand:
                Brand.append(brand.text)
            else:
                Brand.append(np.nan)
                
                
    
            price = i.find("span",class_="o-eZTujG o-byFsZJ o-bkmzIL o-bVSleT")
            if price:
                Price.append(price.text)
            else:
                Price.append(np.nan)  
                
                
            
            total_rating=i.find('span',class_='o-bzsJdq o-fzpimR o-KxopV o-sTQWx o-dThPjR')
            if total_rating:
                text=total_rating.text
                r1=re.findall('([0-9]+)\sRatings',text)
                if r1:
                    Rating_members.append(r1[0])
                else:
                    Rating_members.append(np.nan) 
            else:
                Rating_members.append(np.nan)
                
                

            rating = i.find("p",class_="o-frVjwE o-bdcqVx o-cKuOoN o-lIIwF o-eZTujG")
            if rating:
                text=rating.text
                r1=re.findall('([0-9]\.[0-9])',text)
                if r1:
                    Rating.append(r1[0])
                else:
                    Rating.append(np.nan)
            else:
                Rating.append(np.nan)
                
                
    
            content=i.find('div',class_='o-bqHweY o-bVSleT o-bwCunT o-bfyaNx o-bNxxEB o-fzpihx')
            if content:
                text=content.text
                r1=re.findall('(.+[0-9])\scc',text)
                if r1:
                    CC.append(r1[0])
                else:
                    CC.append(np.nan)
            else:
                    CC.append(np.nan) 
                    
                    
                    
            content=i.find('div',class_='o-bqHweY o-bVSleT o-bwCunT o-bfyaNx o-bNxxEB o-fzpihx')
            if content:
                text=content.text
                r1=re.findall('([0-9].)\skmpl',text)
                if r1:
                    KMPL.append(r1[0])
                else:
                    KMPL.append(np.nan)
            else:
                    KMPL.append(np.nan) 
                    
                    
           
            content=i.find('div',class_='o-bqHweY o-bVSleT o-bwCunT o-bfyaNx o-bNxxEB o-fzpihx')
            if content:
                text=content.text
                r1=re.findall('([\d.]+)\sbhp',text)
                if r1:
                    BHP.append(r1[0])
                else:
                    BHP.append(np.nan)
            else:
                    BHP.append(np.nan) 
                    
                    
                    
            content=i.find('div',class_='o-bqHweY o-bVSleT o-bwCunT o-bfyaNx o-bNxxEB o-fzpihx')
            if content:
                text=content.text
                r1=re.findall('([0-9.]+)\skg',text)
                if r1:
                    Weight.append(r1[0])
                else:
                    Weight.append(np.nan)
            else:
                    Weight.append(np.nan)                  
```

## Plot 6: Outlier Detection


```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Rating', y='Price', data=df)
plt.title('Scatter Plot of Ratings vs Gross Income (Outlier Detection)')
plt.xlabel('Rating')
plt.ylabel('Gross Income (Price)')
# Annotate outliers
for index, row in df.iterrows():
    if row['Rating'] > 4.5 and row['Price'] > 100000:
        plt.text(row['Rating'], row['Price'], row['Bike_Names'], fontsize=9)
plt.grid(True)
plt.show()
```

   #### Identification of outliers in ratings and gross income may highlight exceptional cases worth investigating further. For instance, outlier bikes with exceptionally high ratings and low prices could represent hidden gems in the market, while outliers with low ratings and high prices may indicate areas for improvement or overpricing.
